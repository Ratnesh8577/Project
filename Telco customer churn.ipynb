Telco customer churn refers to the rate at which customers stop doing business with a telecommunications company over a given period of time. In simpler terms, it measures how many customers cancel their services, such as mobile, internet, or cable TV, and either switch to a competitor or stop using the service altogether.

There are two main types of churn:

Voluntary churn: When customers actively decide to leave, often due to factors like poor service quality, high prices, or better offers from competitors.
Involuntary churn: When the service provider ends the relationship, usually because of non-payment or policy violations.

```python
grouped = df.groupby('Senior Citizen')['Churn Label'].value_counts(normalize=True).unstack() * 100

# Plot
grouped.plot(kind='bar', stacked=True, figsize=(5, 5), colormap='Set2')

# Add labels
for i, total in enumerate(grouped.values):
    for j, value in enumerate(total):
        plt.text(i, sum(total[:j+1]) - value/2, f'{value:.1f}%', ha='center', va='center', color='black', fontsize=10)

# Labels and title
plt.title("Churn by Senior Citizen (Percentage)", fontsize=14)
plt.xlabel("Senior Citizen", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
```
